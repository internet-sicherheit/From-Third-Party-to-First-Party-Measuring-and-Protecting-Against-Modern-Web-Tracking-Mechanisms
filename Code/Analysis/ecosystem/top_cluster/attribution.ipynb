```python
import os
import pandas as pd
from tqdm import tqdm

tqdm.pandas()

df = pd.read_csv(os.path.join(os.getcwd(), 'script_url_etld_cluster_1.csv'))
df.head(10)
```

```python
def contains_key_word(url):
    return "google" in url

df = df.dropna()
df['related_to_google'] = df['s_u'].progress_apply(contains_key_word)
df.head(10)
```
