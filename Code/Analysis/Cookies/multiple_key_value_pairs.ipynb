```python
import os
import pandas as pd
from glob import glob

PATH_TO_DATA = os.path.join(os.getcwd(), '..', '..', '..', '02_Data', 'multiple_key_value_pairs', 'values')
FILES = glob(os.path.join(PATH_TO_DATA, '*.csv'))
df = pd.read_csv(FILES[0])
df.head()
```

```python
from urllib.parse import unquote
from tqdm import tqdm
tqdm.pandas()

def url_decode(cookie_value):
    try:
        return unquote(cookie_value)
    except:
        return ""

df['unquote'] = df['value'].progress_apply(url_decode)
df.head()
```

```python
def is_decoded(row):
    return row['unquote'] != row['value']

df['is_decoded'] = df.apply(is_decoded, axis=1)
df.head()
```

```python
print("Multiple_Key_Value_Pairs.1.0")
df['is_decoded'].value_counts()
```
