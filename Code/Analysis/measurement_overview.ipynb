## 1.0: Distinct eTLDS+
Browser_ID | Number of Distinct eTLD+1
-------- | --------
EU1   | 37623
EU2   | 37600
US1   | 31409
US2   | 38262

```python
import numpy as np

# Data
d = {'eu1': 37623, 'eu2': 37600, 'us1': 31409, 'us2': 38262}
values = list(d.values())

# Calculations
min_val = min(values)
max_val = max(values)
avg_val = sum(values) / len(values)
sd_val = np.std(values)

# Output
print(f"[measurement-dataset-overview.1.0]: mean: {avg_val:.0f}, min: {min_val}, max: {max_val}, SD: {sd_val:.0f}")
```
