```python
import re
from simhash2 import Simhash

s1 = "ABCDEF"
s2 = "ABCXDEF"
s3 = "ABCYDEF"

simhashes = [s1, s2, s3]

def get_features(s):
    width = 3 # n_gram = 3
    s = s.lower()
    s = re.sub(r'[^\w]+', '', s)
    return [s[i:i + width] for i in range(max(len(s) - width + 1, 1))]

for s in simhashes:
    print(s, ":", Simhash(get_features(s)).value)
```
