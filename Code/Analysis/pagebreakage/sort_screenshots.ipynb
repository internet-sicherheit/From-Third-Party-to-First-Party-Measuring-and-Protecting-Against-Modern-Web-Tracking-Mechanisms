```python
import os

from glob import glob
from tqdm import tqdm

basepath = os.path.join(os.getcwd(), '..', '..', '..', '02_Data', 'pbdata')
path_w_extension = os.path.join(basepath, 'screenshots_w_extension')
path_wo_extension = os.path.join(basepath, 'screenshot_wo_extension')
path_to_analysis = os.path.join(basepath, 'Screenshots_to_Analyse')

filenames_screenshots_w_extension = glob(os.path.join(path_w_extension, '*.png'))
filenames_screenshots_wo_extension = glob(os.path.join(path_wo_extension, '*.png'))

#print(len(filenames_screenshots_w_extension))
#print(len(filenames_screenshots_wo_extension))

etlds_w_extension = set()
etlds_wo_extension = set()

def extract_sitenames(filename):
    filename = os.path.basename(filename)
    filename_wo_extension = os.path.splitext(filename)[0]
    return filename_wo_extension.split('_')[1]

for filename in tqdm(filenames_screenshots_wo_extension, total=len(filenames_screenshots_wo_extension)):
    etlds_wo_extension.add(extract_sitenames(filename))

for filename in tqdm(filenames_screenshots_w_extension, total=len(filenames_screenshots_w_extension)):
    f = extract_sitenames(filename)
    etlds_w_extension.add(f.replace("_analysis", ''))

intersection = etlds_wo_extension.intersection(etlds_w_extension)
len(intersection)
```

```python
import shutil

for file in tqdm(filenames_screenshots_w_extension, total=len(filenames_screenshots_w_extension)):
    f = extract_sitenames(file).replace("_analysis", '')
    if f in intersection:
        _file = os.path.basename(file).split("_")
        shutil.copyfile(file, os.path.join(path_to_analysis, str(_file[1] + "_" + _file[2])))
```

```python
for file in tqdm(filenames_screenshots_wo_extension, total=len(filenames_screenshots_wo_extension)):
    f = extract_sitenames(file)
    #print(f)
    if f in intersection:
        _file = os.path.basename(file).split("_")[1]
        shutil.copyfile(file, os.path.join(path_to_analysis, _file))
```

```python
n = 1
for i in os.listdir(path_to_analysis):
    if "_analysis" not in i:
        print(n, i)
        n += 1
```
